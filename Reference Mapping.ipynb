```r
library(Seurat)
library(SeuratData)
library(ggplot2)
library(patchwork)
library(stringr)
```

```r
InstallData("pbmcsca")
```

```r
WT_PATH = "data/WT_TRAF6_10X/filtered_gene_bc_matrices/mm10"
KO_PATH = "data/KO_TRAF6_10X/filtered_gene_bc_matrices/mm10"
wt_data = Read10X(data.dir = WT_PATH)
ko_data = Read10X(data.dir = KO_PATH)
rownames(wt_data) = str_to_upper(rownames(wt_data))
rownames(ko_data) = str_to_upper(rownames(ko_data))
wt = CreateSeuratObject(counts = wt_data, project = "wt", min.cells = 3, min.features = 200)
ko = CreateSeuratObject(counts = ko_data, project = "ko", min.cells = 3, min.features = 200)
```

```r
data("pbmcsca")
pbmc.list <- SplitObject(pbmcsca, split.by = "Method")
```

```r
ref = pbmc.list$`10x Chromium (v3)`
```

```r
#rownames(ref$RNA@counts) = str_to_sentence(rownames(ref$RNA@counts))
```

```r
dataset.list = list(wt = wt, ko = ko, ref = ref)
names(dataset.list)
```

```r
for (i in names(dataset.list)) {
    dataset.list[[i]] <- SCTransform(dataset.list[[i]], verbose = FALSE)
}
```

```r
dataset.features <- SelectIntegrationFeatures(object.list = dataset.list, nfeatures = 3000)
dataset.list <- PrepSCTIntegration(object.list = dataset.list, anchor.features = dataset.features)
```

```r
reference_dataset <- which(names(dataset.list) == "ref")
```

```r
dataset.anchors <- FindIntegrationAnchors(object.list = dataset.list, normalization.method = "SCT", anchor.features = dataset.features, reference = reference_dataset)
dataset.integrated <- IntegrateData(anchorset = dataset.anchors, normalization.method = "SCT")
```

```r
dataset.integrated <- RunPCA(object = dataset.integrated, verbose = FALSE)
```

```r
dataset.integrated <- RunUMAP(object = dataset.integrated, dims = 1:15)
```

```r
DimPlot(dataset.integrated, reduction = "pca", group.by = "orig.ident")
DimPlot(dataset.integrated, reduction = "umap", group.by = "orig.ident")
```

```r
DimPlot(dataset.integrated, reduction = "pca", group.by = "CellType")
DimPlot(dataset.integrated, reduction = "umap", group.by = "CellType")
```

```r
DefaultAssay(dataset.integrated) <- "RNA"
# Normalize RNA data for visualization purposes
dataset.integrated <- NormalizeData(dataset.integrated, verbose = FALSE)
FeaturePlot(dataset.integrated, c("CCR7", "S100A4", "GZMB", "GZMK", "GZMH", "TCL1A"))
```
